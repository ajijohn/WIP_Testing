```r
library(terra)
library(sf)
library(tidyterra)
library(MultiscaleDTM)
library(randomForest)
library(caret)
library(vip)
library(rstac)


s_obj <- stac("https://planetarycomputer.microsoft.com/api/stac/v1")
```

```r
wa_diss <- vect("data/wa_shp_diss.gpkg")
```

```r
wa_dem_diss <- rast("data/wa_dem_diss_prj3740.tif")
```

```r
plot(wa_dem_diss)
```

```r
rand_wa_upl <- spatSample(wa_dem_diss, 50000, as.points = TRUE, xy = TRUE)
```

```r
WA_nwi <- vect("data/WA_shapefile_wetlands/WA_Wetlands.shp") |> 
    filter(WETLAND_TY != "Estuarine and Marine Deepwater",
          WETLAND_TY != "Estuarine and Marine Wetland",
          WETLAND_TY != "Other")
```

```r
unique(WA_nwi$WETLAND_TY)
```

```r
wa_upl_pts <- terra::mask(rand_wa_upl, WA_nwi, inverse = T) |> 
    mutate(class = "UPL") |>
    select(class) |> 
    mask(wa_diss) 
```

```r
wa_upl_pts
```

```r
plot(wa_upl_pts)
```

```r
wa_wet_pts <- 
    spatSample(WA_nwi, size = 10000, method = "regular") |> 
    mutate(class = "WET") |>
    select(class) |>
    terra::project("EPSG:3740")
wa_wet_pts
```

```r
wa_all_pts <- rbind(wa_wet_pts, wa_upl_pts)
```

```r
wa_all_pts
```

```r
writeVector(wa_all_pts, "data/wa_all_pts.gpkg", overwrite = T)
```

```r
plot(vect("data/wa_all_pts.gpkg"), "class", type = "classes", cex = 0.5)
```

```r
plot(wa_all_pts[wa_all_pts$class == "WET"])
```

Make the WA HUC8 watershed vector layer with a dataframe that has a row for each HUC8 watershed and a column for each importance metric.

|Name|Imp_Var_1|Imp_Var_2|Imp_Var_3|
|----|---------|---------|---------|
|HUC_Name|NDVI|DEV3|SLP|

```r
wa_hucs_abb <- wa_hucs |> mutate(name_abb = str_replace_all(name, "[^[:alnum:]]", "")) |> select(name_abb)
wa_hucs_abb
```

```r
test_hucs <- wa_hucs_abb[1:5]
```

```r
wip_imp_func <- function(hucs, path) {
    empty_mat <- matrix(nrow = length(hucs), ncol = 4)
    for (i in 1:length(hucs)){
        nameClean <- str_replace_all(paste0(values(hucs[i])), "[^[:alnum:]]", "")
        #print(nameClean)
        mod <- get(load(as.list(list.files(path = paste0(path, "/RF_Models"), 
                                         full.names = TRUE, 
                                         pattern = paste0(nameClean, "rf_model.RData"),
                                         include.dirs = FALSE))[[1]]))
        mod_vi <- vip::vi(mod)
        empty_mat[i, 1] <- nameClean
        empty_mat[i, 2] <- mod_vi[[2, 1]]
        empty_mat[i, 3] <- mod_vi[[3, 1]]
        empty_mat[i, 4] <- mod_vi[[4, 1]]
    }
    vi_df <- as.data.frame(empty_mat)
    names(vi_df) <- c("name_abb", "Imp_Var_1", "Imp_Var_2","Imp_Var_3")
    vi_vect <- dplyr::full_join(hucs, vi_df, by = join_by("name_abb"))
    return(vi_vect)

    #test_df <- as.data.frame(empty_mat)
    #return(test_df)
}

test <- wip_imp_func(test_hucs, datapath)
values(test)
```

```r
wa_hucs_imp <- wip_imp_func(wa_hucs_abb, datapath)
```

```r
wa_hucs_imp_cat <- wa_hucs_imp |> tidyterra::mutate(category1 = case_when(str_detect(Imp_Var_1, "median") ~ "spectral",
                                                      str_detect(Imp_Var_1, "wa_") ~ "terrain",
                                                      str_detect(Imp_Var_1, "NASADEM") ~ "terrain",
                                                      str_detect(Imp_Var_1, "PRISM") ~ "climate",
                                                      str_detect(Imp_Var_1, "canopy") ~ "spectral",
                                                      .default = "other1")) |> 
                                  tidyterra::mutate(category2 = case_when(str_detect(Imp_Var_2, "median") ~ "spectral",
                                                      str_detect(Imp_Var_2, "wa_") ~ "terrain",
                                                      str_detect(Imp_Var_2, "NASADEM") ~ "terrain",
                                                      str_detect(Imp_Var_2, "PRISM") ~ "climate",
                                                      str_detect(Imp_Var_2, "canopy") ~ "spectral",
                                                      .default = "other2")) |> 
                                  tidyterra::mutate(category3 = case_when(str_detect(Imp_Var_3, "median") ~ "spectral",
                                                      str_detect(Imp_Var_3, "wa_") ~ "terrain",
                                                      str_detect(Imp_Var_3, "NASADEM") ~ "terrain",
                                                      str_detect(Imp_Var_3, "PRISM") ~ "climate",
                                                      str_detect(Imp_Var_3, "canopy") ~ "spectral",
                                                      .default = "other3"))
```

```r
plot(wa_hucs_imp_cat, "category1", main = "First Important")
plot(wa_hucs_imp_cat, "category2", main = "Second Important")
plot(wa_hucs_imp_cat, "category3", main = "Third Important")
```
